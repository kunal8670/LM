## Data Cleaning and Preparation

###  Importing Req Lib

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

###  Loading Dataset

```python
df = pd.read_csv("/content/drive/MyDrive/User/Student_Academic_Performance.csv")
print("Initial Dataset Shape:", df.shape)
print("\nFirst 5 Records of Dataset:\n")
print(df.head())
```

###  Check Missing Values

```python
print("\nMissing Values Before Cleaning:\n")
print(df.isnull().sum())
```

###  Handle Missing Values

```python
df["Study_Hours"] = df["Study_Hours"].fillna(df["Study_Hours"].mean())
df["Attendance_Percentage"] = df["Attendance_Percentage"].fillna(df["Attendance_Percentage"].mean())
df["Internal_Assessment"] = df["Internal_Assessment"].fillna(df["Internal_Assessment"].mean())
print("\nMissing Values After Cleaning:\n")
print(df.isnull().sum())
```

### Removeing Duplicate Records

```python
print("\nDataset Shape Before Removing Duplicates:", df.shape)
df = df.drop_duplicates()
print("Dataset Shape After Removing Duplicates:", df.shape)
```

### Removing Irreleveant Column

```python
print("\nColumns Before Removing Department Column:\n", df.columns)
df = df.drop("Department", axis=1)
print("\nColumns After Removing Department Column:\n", df.columns)
```

###  Encoding Categorical Variables

```python
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
df["Parent_Education"] = le.fit_transform(df["Parent_Education"])
df["ESE_Preparation"] = le.fit_transform(df["ESE_Preparation"])
print("\nDataset After Encoding Categorical Variables:\n")
print(df.head())
```

### Create Target Variable

```python
df["Average_Score"] = (df["Internal_Assessment"] +
                        df["Lab_Test_Score"] +
                        df["External_POE_Score"]) / 3
df["Result"] = df["Average_Score"].apply(lambda x: "Pass" if x >= 15 else "Fail")
print("\nDataset with Result Column:\n")
print(df[["Internal_Assessment","Lab_Test_Score","External_POE_Score","Average_Score","Result"]].head())
```

### Feature Scaling

```python
scaler = StandardScaler()
features = ["Study_Hours",
            "Attendance_Percentage",
            "Internal_Assessment",
            "Lab_Test_Score",
            "External_POE_Score"]
print("\nValues Before Scaling:\n")
print(df[features].head())
df[features] = scaler.fit_transform(df[features])
print("\nValues After Scaling:\n")
print(df[features].head())
```

###  Separating Features and Target

```python
if "Average_Score" not in df.columns:
    df["Average_Score"] = (df["Internal_Assessment"] +
                            df["Lab_Test_Score"] +
                            df["External_POE_Score"]) / 3

if "Result" not in df.columns:
    df["Result"] = df["Average_Score"].apply(lambda x: "Pass" if x >= 15 else "Fail")


X = df.drop(["Result","Average_Score"], axis=1)
y = df["Result"]
print("\nFeature Matrix Shape:", X.shape)
print("Target Vector Shape:", y.shape)
```
