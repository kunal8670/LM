## Experiment 9: Implementation of K-Means Clustering Algorithm on Iris Dataset and Visualization of Clusters

### Part A: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### Part B: Load Dataset

```python
data = pd.read_csv("iris.data.csv", header=None)
data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
print("First 5 rows:")
print(data.head())
print("\nColumn Names:")
print(data.columns)
print("\nDataset Shape:")
print(data.shape)
```

### Part C: Select Features

```python
X = data.iloc[:, :-1]
print("\nSelected Features:")
print(X.head())
```

### Part D: Elbow Method

```python
wcss = []
for i in range(1, 11):
    kmeans_temp = KMeans(n_clusters=i, random_state=42)
    kmeans_temp.fit(X)
    wcss.append(kmeans_temp.inertia_)

print("\nWCSS Values:")
print(wcss)
```

### Part E: Plot Elbow Graph

```python
plt.figure()
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.show()
```

### Part F: Train K-Means Model

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)
print("\nK-Means Model Trained Successfully")
```

### Part G: Add Cluster Labels

```python
data['Cluster'] = clusters
print("\nDataset with Cluster Labels:")
print(data.head())
```

### Part H: Visualization of Clusters

```python
plt.figure()
for i in range(3):
    plt.scatter(
        X.iloc[clusters == i, 0],
        X.iloc[clusters == i, 1],
        label=f'Cluster {i+1}'
    )
centers = kmeans.cluster_centers_
plt.scatter(
    centers[:, 0],
    centers[:, 1],
    c='red',
    marker='x',
    s=200,
    label='Centroids'
)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("K-Means Clustering (Iris Dataset)")
plt.legend()
plt.show()
```
