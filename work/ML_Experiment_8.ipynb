## Experiment 8: Implementation of Naive Bayes Classification Algorithm for Mushroom Dataset

### Part A: Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score, classification_report
```

### Part B: Load Dataset

```python
data = pd.read_csv("mushrooms.csv")

print("First 5 rows:")
print(data.head())

print("\nColumn Names:")
print(data.columns)

print("\nDataset Shape:")
print(data.shape)
```

### Part C: Encode Categorical Data

```python
le = LabelEncoder()

for col in data.columns:
    data[col] = le.fit_transform(data[col])

print("\nAfter Encoding:")
print(data.head())

print("\nTarget Encoding: 0 = Edible, 1 = Poisonous")
```

### Part D: Define Features and Target

```python
X = data.drop('class', axis=1)
Y = data['class']

print("\nFeatures (X):")
print(X.head())

print("\nTarget (Y):")
print(Y.head())
```

### Part E: Train-Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

print("\nTraining Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

### Part F: Train Naive Bayes Model

```python
nb_model = CategoricalNB()

nb_model.fit(X_train, Y_train)

print("\nNaive Bayes Model Trained Successfully")
```

### Part G: Prediction

```python
y_pred = nb_model.predict(X_test)

print("\nFirst 10 Predictions:")
print(y_pred[:10])
```

### Part H: Evaluation

```python
accuracy = accuracy_score(Y_test, y_pred)

print("\nAccuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(Y_test, y_pred))
```

### Part I: Predict New Data

```python
new_data = X.iloc[0:1]

print("\nNew Data Sample:")
print(new_data)

prediction = nb_model.predict(new_data)

print("\nNew Data Prediction (0=Edible, 1=Poisonous):", prediction[0])
```
