```python
import numpy as np
```

#### NumPy Array

```python
arr = np.array([12, 18, 25, 30, 45, 50])
print("Array:", arr)
```

####  Numerical Operations

```python
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Standard Deviation:", np.std(arr))
print("Maximum Value:", np.max(arr))
print("Minimum Value:", np.min(arr))
```

#### Two-Dimensional Array

```python
matrix = np.array([[1, 2, 3],
                    [4, 5, 6]])

print("Matrix:\n", matrix)
print("Matrix Shape:", matrix.shape)
print("Transpose:\n", matrix.T)
```

```python
import pandas as pd
```

#### DataFrame

```python
data = {
    "Name": ["Amit", "Rahul", "Sneha", "Pooja"],
    "Marks": [85, 90, 88, 76],
    "Age": [20, 21, 19, 22]
}

df = pd.DataFrame(data)
print(df)
```

####  Inspect  DataFrame

```python
print("Shape:", df.shape)
print("Columns:", df.columns)
print("Data Types:\n", df.dtypes)
print("Summary Statistics:\n", df.describe())
```

#### Filtering and Column Addition

```python
print("Students with Marks > 80:\n", df[df["Marks"] > 80])

df["Grade"] = ["A", "A", "A", "B"]
print("Updated DataFrame:\n", df)
```

###  Scikit learn Dataset Expo

#### Load Built-in Dataset

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

#### Dataset to Pandas DataFrame

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df["target"] = iris.target
```

#### print Dataset

```python
print(iris_df.head())
```

####  Dataset Shape

```python
print("Dataset Shape:", iris_df.shape)
```

####  Dataset Info

```python
iris_df.info()
print("Summary Statistics:\n", iris_df.describe())
```

#### Separate Features and Target

```python
X = iris_df.iloc[:, :-1]
y = iris_df.iloc[:, -1]

print("Feature Matrix Shape:", X.shape)
print("Target Vector Shape:", y.shape)
```
