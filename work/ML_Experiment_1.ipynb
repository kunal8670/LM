## Experiment 1: Introduction to Python Machine Learning Libraries (NumPy, Pandas, Scikit-learn)

### Part A: NumPy Operations

#### 1. Import Required Library

```python
import numpy as np
```

#### 2. Create a NumPy Array

```python
arr = np.array([12, 18, 25, 30, 45, 50])
print("Array:", arr)
```

#### 3. Perform Numerical Operations

```python
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Standard Deviation:", np.std(arr))
print("Maximum Value:", np.max(arr))
print("Minimum Value:", np.min(arr))
```

#### 4. Create a Two-Dimensional Array (Matrix)

```python
matrix = np.array([[1, 2, 3],
                    [4, 5, 6]])

print("Matrix:\n", matrix)
print("Matrix Shape:", matrix.shape)
print("Transpose:\n", matrix.T)
```

### Part B: Pandas Data Handling

#### 1. Import Pandas

```python
import pandas as pd
```

#### 2. Create a DataFrame

```python
data = {
    "Name": ["Amit", "Rahul", "Sneha", "Pooja"],
    "Marks": [85, 90, 88, 76],
    "Age": [20, 21, 19, 22]
}

df = pd.DataFrame(data)
print(df)
```

#### 3. Inspect the DataFrame

```python
print("Shape:", df.shape)
print("Columns:", df.columns)
print("Data Types:\n", df.dtypes)
print("Summary Statistics:\n", df.describe())
```

#### 4. Perform Basic Filtering and Column Addition

```python
print("Students with Marks > 80:\n", df[df["Marks"] > 80])

df["Grade"] = ["A", "A", "A", "B"]
print("Updated DataFrame:\n", df)
```

### Part C: Scikit-learn Dataset Exploration

#### 1. Import and Load Built-in Dataset

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

#### 2. Convert Dataset into Pandas DataFrame

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df["target"] = iris.target
```

#### 3. Display Dataset

```python
print(iris_df.head())
```

#### 4. Check Dataset Shape

```python
print("Dataset Shape:", iris_df.shape)
```

#### 5. Display Dataset Information

```python
iris_df.info()
print("Summary Statistics:\n", iris_df.describe())
```

#### 6. Separate Features and Target

```python
X = iris_df.iloc[:, :-1]
y = iris_df.iloc[:, -1]

print("Feature Matrix Shape:", X.shape)
print("Target Vector Shape:", y.shape)
```
