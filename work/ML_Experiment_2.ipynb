##  Data Visualization using Matplotlib and Seaborn

### import lib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### loding Titanic Dataset

```python
df = pd.read_csv("/content/drive/MyDrive/User/Titanic-Dataset.csv")

print("Dataset Shape:", df.shape)
print(df.head())
```

###  Data Info

```python
print(df.info())
print("Summary Statistics:\n", df.describe())
```

###  Bar Chart / Survival Distribution

```python
plt.figure(figsize=(6,4))
df["Survived"].value_counts().plot(kind="bar")
plt.title("Survival Distribution")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

### CountPlot Survival by Gender

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival Count by Gender")
plt.show()
```

###  Hist for  Age Distribution

```python
plt.figure(figsize=(6,4))
plt.hist(df["Age"].dropna(), bins=15)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

###  Scatter Plot for Age vs Fare

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["Fare"])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

###  Boxplot Fare by Passenger Class

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

###  Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
