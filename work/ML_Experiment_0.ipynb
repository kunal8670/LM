```python
iris = datasets.load_iris()
```

#### Step 4: Convert Iris Dataset into DataFrame

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
print(iris_df.head())
print(iris_df.shape)
iris_df.info()
```
