## Experiment 5: Implementation of Logistic Regression for Binary Classification using Loan Approval Dataset

### Part A: Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

### Part B: Load Dataset

```python
data = pd.read_csv("loan_approval_dataset.csv")
print("First 5 rows:")
print(data.head())

print("\nColumn Names:")
print(data.columns)

print("\nDataset Shape:")
print(data.shape)
```

### Part C: Encode Categorical Data

```python
le = LabelEncoder()

data[' education'] = le.fit_transform(data[' education'])
data[' self_employed'] = le.fit_transform(data[' self_employed'])
data[' loan_status'] = le.fit_transform(data[' loan_status'])

print("After Encoding:")
print(data.head())
```

### Part D: Remove Unnecessary Column

```python
data = data.drop('loan_id', axis=1)

print(data.columns)
```

### Part E: Define Features and Target

```python
X = data.drop(' loan_status', axis=1)
Y = data[' loan_status']
```

### Part F: Train-Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

### Part G: Feature Scaling

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Part H: Train Model

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, Y_train)
```

### Part I: Prediction

```python
Y_pred = model.predict(X_test_scaled)
print(Y_pred[:10])
```

### Part J: Model Evaluation

```python
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
print(confusion_matrix(Y_test, Y_pred))
```

### Part K: Predict New Data

```python
new_data = [[2, 0, 0, 5000000, 15000000, 10, 750, 3000000, 2000000, 10000000, 4000000]]

new_data_df = pd.DataFrame(new_data, columns=X.columns)

new_data_scaled = scaler.transform(new_data_df)

prediction = model.predict(new_data_scaled)
print("Prediction (0=Approved, 1=Rejected):", prediction[0])
```
