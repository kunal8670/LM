## Decision Tree and Random Forest

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("loan_approval_dataset.csv")

print("First 5 rows:")
print(data.head())

print("\nColumn Names:")
print(data.columns)

print("\nDataset Shape:")
print(data.shape)
```

###  Encodeing

```python
le = LabelEncoder()

data[' education'] = le.fit_transform(data[' education'])
data[' self_employed'] = le.fit_transform(data[' self_employed'])
data[' loan_status'] = le.fit_transform(data[' loan_status'])

print("After Encoding:")
print(data.head())
```

###  Removing Unnecessary Column

```python
data = data.drop('loan_id', axis=1)

print(data.columns)
```

###  Define Features and Target

```python
X = data.drop(' loan_status', axis=1)
Y = data[' loan_status']
```

### Train-Test Split





```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

:### Decision Tree using Gini

```python
dt_gini = DecisionTreeClassifier(criterion='gini')

dt_gini.fit(X_train, Y_train)

gini_pred = dt_gini.predict(X_test)

print("Decision Tree (Gini) Predictions:")
print(gini_pred[:10])

gini_acc = accuracy_score(Y_test, gini_pred)

print("Decision Tree (Gini) Accuracy:", gini_acc)
```

## Decision Tree using Entropy

```python
dt_entropy = DecisionTreeClassifier(criterion='entropy')

dt_entropy.fit(X_train, Y_train)

entropy_pred = dt_entropy.predict(X_test)

print("Decision Tree (Entropy) Predictions:")
print(entropy_pred[:10])

entropy_acc = accuracy_score(Y_test, entropy_pred)

print("Decision Tree (Entropy) Accuracy:", entropy_acc)
```

###  Random Forest Classifier

```python
rf_model = RandomForestClassifier(n_estimators=100)

rf_model.fit(X_train, Y_train)

rf_pred = rf_model.predict(X_test)

print("Random Forest Predictions:")
print(rf_pred[:10])

rf_accuracy = accuracy_score(Y_test, rf_pred)

print("Random Forest Accuracy:", rf_accuracy)
```

###  Model Comparison

```python
print("\n--- Model Comparison ---")
print("Decision Tree (Gini):", gini_acc)
print("Decision Tree (Entropy):", entropy_acc)
print("Random Forest:", rf_accuracy)
```

###  Predict for New Data

```python
new_data = [[2, 0, 0, 5000000, 15000000, 10, 750, 3000000, 2000000, 10000000, 4000000]]

new_data_df = pd.DataFrame(new_data, columns=X.columns)

prediction = rf_model.predict(new_data_df)

print("Prediction (0=Approved, 1=Rejected):", prediction[0])
```
